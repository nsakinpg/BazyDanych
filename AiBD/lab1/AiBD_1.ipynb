<h3>Zadanie 3</h3>
Zdefiniowanie funkcji i sporządzenie wykresów dla 3 różnych przedziałów.

$f(x)=x^2+5$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f = lambda x: x**2 + 5

x1 = np.linspace(-1,1)
x2 = np.linspace(-6,6)
x3 = np.linspace(0,5)
```

```python
plt.plot(x1, f(x1))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wykres funkcji f dla -1<x<1')
plt.legend(['f(x)'])
plt.grid()
```

```python
plt.plot(x2, f(x2))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wykres funkcji f dla -6<x<6')
plt.legend(['f(x)'])
plt.grid()
```

```python
plt.plot(x3, f(x3))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wykres funkcji f dla 0<x<5')
plt.legend(['f(x)'])
plt.grid()
```
