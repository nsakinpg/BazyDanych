Dataset dla mnie to -> 10_POMORSKIE.csv

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('laboratorium-6-nsakinpg-main/10_POMORSKIE.csv',index_col=0,parse_dates=[0])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
sns.histplot(data=df, x='Ocena', bins=len(df['Ocena'].value_counts()))
plt.ylabel('Count')
plt.xlabel('Ocena')
plt.title('Rozkład ocen kupujących')
```

Jak widzimy więkość ocen waha się pomiędzy 2 i 3.

```python
sns.histplot(data=df, x='Marka', bins=len(df['Marka'].value_counts()))
plt.ylabel('Count')
plt.xlabel('Marka')
plt.title('Popularność danych marek przez kupujących')
```

Najpopularniejszą marką jest Tefal.

```python
sns.histplot(data=df, x='Płeć kupującego', bins=len(df['Płeć kupującego'].value_counts()))
plt.ylabel('Count')
plt.xlabel('Płeć kupującego')
plt.title('Porównanie płci kupujących')
```

Duża liczby osób nie podała swojej płci.

```python
sns.histplot(data=df, x='Wiek kupującego', bins=len(df['Wiek kupującego'].value_counts()))
plt.ylabel('Count')
plt.xlabel('Wiek kupującego')
plt.title('Rozkład wieku kupujących')
```

Rozkład wieku układ się w miarę normalnie z dziwnym pikiem dla wartości 44 lat.

```python
sns.histplot(data=df, x='Dni od zakupu', bins=len(df['Dni od zakupu'].value_counts()))
plt.ylabel('Count')
plt.xlabel('Dni od zakupu')
plt.title('Dni od zakupu w momencie wystawienia oceny')
```

Rokład dni od zakupu prezentuje się normalnie.
